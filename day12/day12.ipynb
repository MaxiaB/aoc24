```python
file_path = 'input.txt'


def explore_region(grid, x, y, visited):
    stack = [(x, y)]
    region_char = grid[x][y]
    area = 0
    perimeter = 0
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    while stack:
        cx, cy = stack.pop()
        if (cx, cy) in visited:
            continue
        visited.add((cx, cy))
        area += 1

        for dx, dy in directions:
            nx, ny = cx + dx, cy + dy
            if 0 <= nx < len(grid) and 0 <= ny < len(grid[0]):
                if grid[nx][ny] == region_char:
                    if (nx, ny) not in visited:
                        stack.append((nx, ny))
                else:
                    perimeter += 1
            else:
                perimeter += 1

    return area, perimeter

# Main function to process the grid
def calculate_fencing_cost(grid):
    total_cost = 0
    visited = set()

    for x in range(len(grid)):
        for y in range(len(grid[0])):
            if (x, y) not in visited:
                area, perimeter = explore_region(grid, x, y, visited)
                total_cost += area * perimeter

    return total_cost

# Read the grid from the input file
with open(file_path, 'r') as file:
    garden_map = [list(line.strip()) for line in file.readlines()]

# Calculate the total cost of fencing
total_cost = calculate_fencing_cost(garden_map)
total_cost
```

```python
# From LeWagon Community

import numpy as np
from scipy.ndimage import label

# Initialize gridor char
grid = np.array(
    [
        [
            *x.strip()
        ]
        for x
        in open("input.txt")
    ]
)

# Initialize price
price = 0

# Loop over every plant type
for plant in np.unique(
    grid
):
    # Label each individual region for this plant type
    regions, region_count = label(
        np.where(
            grid == plant,
            1,
            0
        )
    )
    # Loop over each region
    for region in range(
        region_count
    ):
        # Mask everything but this region
        unique_region = np.pad(
            np.where(
                regions == region + 1,
                1,
                0
            ),
            1
        )
        # Area is just the count of plots in the region
        area = np.sum(
            unique_region
        )
        # Instead of counting edges, we're counting corners:
        # Loop over every plot, and if it has a plant (first condition)
        # Second condition A: It is an outside (all 3 connecting plots are not the same plant type) or inside (only the diagonal plant type is different) corner
        # Second condition B: It is an outside corner, and the plot across is part of the same region, but not connected
        border = sum(
            1
            for r in range(unique_region.shape[0])
            for c in range(unique_region.shape[1])
            for x in range(2)
            for y in range(2)
            if unique_region[r, c] != 0 and (
                (
                    (
                        np.count_nonzero(
                            np.pad(
                                unique_region,
                                1
                            )[
                                r+x:r+2+x,
                                c+y:c+2+y
                            ] == 0
                        ) in (
                            1,
                            3
                        )
                    ) and np.pad(
                        unique_region,
                        1
                    )[
                        r+2*x,
                        c+2*y
                    ] == 0
                ) or (
                    (
                        np.count_nonzero(
                            np.pad(
                                unique_region,
                                1
                            )[
                                r+x:r+2+x,
                                c+y:c+2+y
                            ] == 0
                        ) == 2
                    ) and np.pad(
                        unique_region,
                        1
                    )[
                        r+2*x,
                        c+2*y
                    ] == 1
                )
            )
        )

        # Add to total price
        price += area * border

# Output
price
```
