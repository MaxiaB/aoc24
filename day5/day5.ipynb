```python
# PART 2
from collections import defaultdict, deque


with open('input.txt', 'r') as f:
    lines = f.read().splitlines()

# Split the input into rules and updates
# Rules are in format "X|Y" where X must come before Y
rules = []
updates = []
is_rule_section = True  # Flag to track which section we're parsing

for line in lines:
    if line.strip() == '':
        is_rule_section = False
        continue
    if is_rule_section:
        rules.append(line.strip())
    else:
        updates.append(line.strip())

# Parse the rules into a set of tuples
rules_set = set()
for rule in rules:
    X, Y = rule.split('|')
    X = int(X)
    Y = int(Y)
    rules_set.add((X, Y))

# Identify incorrectly ordered updates
incorrect_updates = []
for update_line in updates:
    pages = [int(x) for x in update_line.strip().split(',')]
    page_indices = {page: idx for idx, page in enumerate(pages)}
    is_valid = True
    for X, Y in rules_set:
        if X in page_indices and Y in page_indices:
            if page_indices[X] >= page_indices[Y]:
                is_valid = False
                break
    if not is_valid:
        incorrect_updates.append(pages)

# Function to perform topological sort on the pages
def topological_sort(pages, rules_set):
    # Build the adjacency list and in-degree count
    adj = defaultdict(list)
    in_degree = {page: 0 for page in pages}
    pages_set = set(pages)
    for X, Y in rules_set:
        if X in pages_set and Y in pages_set:
            adj[X].append(Y)
            in_degree[Y] += 1
    # Kahn's algorithm for topological sorting
    queue = deque([node for node in pages if in_degree[node] == 0])
    sorted_pages = []
    while queue:
        node = queue.popleft()
        sorted_pages.append(node)
        for neighbor in adj[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
    if len(sorted_pages) != len(pages):
        raise Exception("Cycle detected in the graph; topological sort not possible.")
    return sorted_pages

# Reorder each incorrect update and calculate the middle page numbers
middle_pages = []
for pages in incorrect_updates:
    sorted_pages = topological_sort(pages, rules_set)
    middle_page = sorted_pages[len(sorted_pages) // 2]
    middle_pages.append(middle_page)

# Calculate the total sum of middle pages
total = sum(middle_pages)
print(total)
```
